```python
import numpy as np
from scipy.optimize import minimize

# Define the function to calculate u_total for a given b and parameters
def calc_u_total(b, y3, y1, gamma, beta, R, y_bar, y_underline):
    psi = np.log(y3) - np.log(gamma * y3)
    b_bar = (1 - gamma * np.exp(- beta * psi)) * y_bar
    b_underline = (1 - gamma * np.exp(- beta * psi)) * y_underline
    y_tau = b / (1 - gamma * np.exp(-beta * psi))
    q_tau = (1 / R) * (y_bar - y_tau) / (y_bar - y_underline)

    term1 = np.log(y1 + q_tau * b)
    term2 = (beta / (y_bar - y_underline)) * ((y_bar - b) * np.log(y_bar - b) - (y_tau - b) * np.log(y_tau - b) - (y_bar - y_tau))
    term3 = ((beta) ** 2 / (y_bar - y_underline)) * np.log(y3) * (y_bar - y_tau)
    term4 = (beta / (y_bar - y_underline)) * (y_tau * np.log(gamma * y_tau) - y_underline * np.log(gamma * y_underline) - (y_tau - y_underline))
    term5 = ((beta) ** 2 / (y_bar - y_underline)) * np.log(gamma * y3) * (y_tau - y_underline)
    return term1 + term2 + term3 + term4 + term5

# Modified inner optimization to return the optimal b value along with the maximum u_total
def inner_optimization(params):
    y3, y1, gamma, beta, R, y_bar, y_underline = params
    result = minimize(lambda b: -calc_u_total(b, y3, y1, gamma, beta, R, y_bar, y_underline), x0=[(y_bar + y_underline) / 2], bounds=[(y_underline, y_bar)])
    if result.success:
        optimal_b = result.x[0]
        max_u_total = -result.fun
        return max_u_total, optimal_b
    else:
        return np.nan, None  # Optimization failed

# Outer optimization remains unchanged
def outer_objective(params):
    max_u_total, optimal_b = inner_optimization(params)
    return -max_u_total  # Negate because we are maximizing


# Initial guesses for the parameters
initial_params = [75, 60, 0.5, 0.1, 1.05, 100, 50]  # Example: [y3, y1, gamma, beta, R, y_bar, y_underline]

# Parameter bounds
bounds = [(50, 100), (40, 80), (0.1, 1), (0.01, 0.5), (1, 2), (80, 120), (30, 70)]  # Example bounds

# Run the outer optimization
result = minimize(outer_objective, initial_params, bounds=bounds)

if result.success:
    optimized_params = result.x
    max_u_total, optimal_b = inner_optimization(optimized_params)
    y3, y1, gamma, beta, R, y_bar, y_underline = optimized_params
    psi_optimal = np.log(y3) - np.log(gamma * y3)
    b_bar_optimal = (1 - gamma * np.exp(- beta * psi_optimal)) * y_bar
    b_underline_optimal = (1 - gamma * np.exp(- beta * psi_optimal)) * y_underline

    # Check if optimal_b is within the range
    in_range = b_underline_optimal <= optimal_b <= b_bar_optimal

    # Print optimized parameters with names
    param_names = ["y3", "y1", "gamma", "beta", "R", "y_bar", "y_underline"]
    print("Optimized parameters:")
    for name, value in zip(param_names, optimized_params):
        print(f"{name}: {value:.4f}")

    print(f"Maximum u_total: {max_u_total:.4f}")
    print(f"Optimal b: {optimal_b:.4f}")
    print(f"b_bar (Optimal): {b_bar_optimal:.4f}")
    print(f"b_underline (Optimal): {b_underline_optimal:.4f}")
    print(f"The optimal b is within the range of b_bar to b_underline: {'Yes' if in_range else 'No'}")
else:
    print("Optimization failed.")
```
