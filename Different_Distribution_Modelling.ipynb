```python
import numpy as np
from scipy.integrate import quad
import matplotlib.pyplot as plt

# Assign hypothetical values to the parameters
y1 = 10
beta = 0.9
y_bar = 50
y_underline = 10
y3 = 10
R = 1
gamma = 0.95

# Define a function for y^tau(b)
def y_tau_b_function(b, gamma, beta, y3):
    return b / (1 - gamma * np.exp(-1 * beta * (np.log(y3) - np.log(gamma * y3))))

# Define the function for q as a function of y^tau(b)
def q_function(y_tau_b, R, y_bar, y_underline):
    return (1 / R) * (y_bar - y_tau_b) / (y_bar - y_underline)

# Define the function for the first part of the equation
def first_part(y1, q, b):
    return np.log(y1 + q * b)

# Define the functions for the second and third parts of the integral
def second_part(y2, b, beta, y3):
    return np.log(y2 - b) + beta * np.log(y3)

def third_part(y2, gamma, beta, y3):
    return np.log(gamma * y2) + beta * np.log(gamma * y3)

# Function to calculate U for a given b
def calculate_U(b, gamma, beta, y3, R, y1, y_bar, y_underline):
    y_tau_b = y_tau_b_function(b, gamma, beta, y3)
    q = q_function(y_tau_b, R, y_bar, y_underline)
    integral_1, _ = quad(second_part, y_tau_b, y_bar, args=(b, beta, y3))
    integral_2, _ = quad(third_part, y_underline, y_tau_b, args=(gamma, beta, y3))
    U = first_part(y1, q, b) + (beta / (y_bar - y_underline)) * integral_1 + (beta / (y_bar - y_underline)) * integral_2
    return U

# Calculate b_bar and b_underline
b_bar = y_bar * (1 - gamma * np.exp(-1 * beta * (np.log(y3) - np.log(gamma * y3))))
b_underline = y_underline * (1 - gamma * np.exp(-1 * beta * (np.log(y3) - np.log(gamma * y3))))

# Generate b values
b_values = np.linspace(b_underline, b_bar, 100)

# Calculate U for each b value
U_values = [calculate_U(b, gamma, beta, y3, R, y1, y_bar, y_underline) for b in b_values]

# Plotting
plt.plot(b_values, U_values, label='U vs b')
plt.xlabel('b')
plt.ylabel('Utility (U)')
plt.title('Utility Function U as a function of b')
plt.grid(True, linestyle='--')
plt.show()
```
